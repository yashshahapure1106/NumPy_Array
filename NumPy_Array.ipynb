### Normal Distribution

```python
heights = np.random.normal(loc=5.5, scale=1/3, size=1000)
import seaborn as sns
sns.histplot(heights, kde=True)
```

```python
np.random.seed(42)
heights = np.random.normal(loc=5.5, scale=1/3, size=1000)
import seaborn as sns
sns.histplot(heights, kde=True)
```
